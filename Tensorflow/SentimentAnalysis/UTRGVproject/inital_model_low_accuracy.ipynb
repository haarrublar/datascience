One recommended process before stemming is called normalization, and we do it applying the lower case function to all the data such that words as _The_ and _the_ will not be different.

### Normalization

```python
list1 = [data[i].lower() for i in range(len(data))]
list1
```

### Tokenization.

Tokenization is the language process to break up the strings into words and punctuation. A token is the technical name for a  sequence of characters that we want to treat as a group.

We start doing this process using the normalized list.

```python
gen_token = [nltk.word_tokenize(list1[i]) for i in range(len(list1))]
```

```python
gen_token[4]
```

### Stopping words Algorihtm

There is also a corpus of stopwords, that is, high-frequency words such as the, to, and also that we sometimes want to filter out of a document before further processing. Stopwords usually have little lexical content, and their presence in a text fails to distinguish it from other texts.

```python
from nltk.corpus import stopwords

# stoppign words for english
genen_token = []
for word in gen_token:
    if word not in stopwords.words('english'):
        genen_token.append(word)
        
# stoppign words for all stopwords
for word in gen_token:
    if word in stopwords.words():
        gen_token.remove(word)
```

```python
genen_token
```

Apparently, both lists are the same.

### Stemmers.

NLTK includes several off-the-shelf stemmers, and if you ever need a stemmer, you should use one of these in preference to crafting your own using regular expressions, since NLTK’s stemmers handle a wide range of irregular cases.
